# Covariance

```python
import numpy as np

# Sample data:
# Variable X: Hours studied per week
hours_studied = np.array([2, 4, 5, 7, 8, 10, 11, 13, 15, 16])
# Variable Y: Exam score (out of 100)
exam_scores = np.array([55, 60, 62, 71, 75, 80, 83, 88, 91, 95])
```

```python
# np.cov() takes a 2D array where each row is a variable.
# We stack our two 1D arrays as rows in a 2D array.
data = np.stack((hours_studied, exam_scores), axis=0)

# The function will calculate sample covariance by default (dividing by n-1).
covariance_matrix = np.cov(data)

print(f"Hours Studied: {hours_studied}")
print(f"Exam Scores:   {exam_scores}")
print("\n" + "="*40 + "\n")

print("Covariance Matrix")
print(covariance_matrix)

# The matrix is a 2x2 matrix because we have 2 variables.
# It's structured as:
# [[ Cov(X,X), Cov(X,Y) ],
#  [ Cov(Y,X), Cov(Y,Y) ]]
```

```python
# The value on the diagonal at [0, 0] is the variance of hours_studied.
var_hours = covariance_matrix[0, 0]
print(f"Variance of Hours Studied (Cov(X,X)): {var_hours:.2f}")

# The value on the diagonal at [1, 1] is the variance of exam_scores.
var_scores = covariance_matrix[1, 1]
print(f"Variance of Exam Scores (Cov(Y,Y)): {var_scores:.2f}")

# The off-diagonal values are the covariance between the two variables.
# Cov(X,Y) is always equal to Cov(Y,X).
cov_xy = covariance_matrix[0, 1]
print(f"\nCovariance between Hours and Scores (Cov(X,Y)): {cov_xy:.2f}")
print("Since the covariance is positive, it indicates that as hours studied increase, exam scores tend to increase.")
```

# Correlation

```python
import numpy as np
import matplotlib.pyplot as plt

# Positive Correlation
# Data: Ice cream sales vs. temperature
temperature = np.array([14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2])
ice_cream_sales = np.array([215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408])

# Calculate the correlation matrix
# np.corrcoef expects rows to be variables, so we don't need to stack them.
positive_corr_matrix = np.corrcoef(temperature, ice_cream_sales)

print("Positive Correlation: Temperature vs. Ice Cream Sales")
print("Correlation Matrix:")
print(positive_corr_matrix)
```

```python
# The off-diagonal element [0, 1] is the correlation coefficient
r_positive = positive_corr_matrix[0, 1]
print(f"\nCorrelation Coefficient (r): {r_positive:.4f}")
print("This value is very close to +1, indicating a strong positive linear relationship.")
```

```python
# Negative Correlation
# Data: Hours spent playing video games vs. GPA
game_hours = np.array([0, 0, 1, 1.5, 2, 2, 3, 3.5, 4, 5, 5.5, 6])
gpa = np.array([4.0, 3.8, 3.5, 3.6, 3.2, 3.0, 2.8, 2.5, 2.2, 1.8, 2.0, 1.5])

# Calculate the correlation matrix
negative_corr_matrix = np.corrcoef(game_hours, gpa)

print("Negative Correlation: Video Game Hours vs. GPA")
print("Correlation Matrix:")
print(negative_corr_matrix)
# The off-diagonal element [0, 1] is the correlation coefficient
r_negative = negative_corr_matrix[0, 1]
print(f"\nCorrelation Coefficient (r): {r_negative:.4f}")
print("This value is very close to -1, indicating a strong negative linear relationship.")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.scatter(temperature, ice_cream_sales, color='red')
ax1.set_title(f'Positive Correlation (r={r_positive:.2f})')
ax1.set_xlabel('Temperature (Â°C)')
ax1.set_ylabel('Ice Cream Sales ($)')

ax2.scatter(game_hours, gpa, color='blue')
ax2.set_title(f'Negative Correlation (r={r_negative:.2f})')
ax2.set_xlabel('Hours of Video Games per Day')
ax2.set_ylabel('GPA')

plt.tight_layout()
plt.show()
```
